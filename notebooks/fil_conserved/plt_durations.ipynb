```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
def duration_m(d, slope=1, clip=None):
    if d < round(365*1.5):
        return 1
    else:
        y1_slp1 = 1
        y2_slp1 = 2 - 183/365.
        x1_slp1 = round(365*1.5)
        x2_slp1 = 365*2
        m = (y2_slp1 - y1_slp1) / (x2_slp1 - x1_slp1)
        m *= slope
        # y-y1 = m*(x-x1)
        y = m*(d - x1_slp1) + y1_slp1
        if clip is not None:
            if y > clip:
                y = clip
        return y
```

```python
days = np.arange(0,365*5)
slp_1 = np.zeros(len(days))
slp_0_1 = np.zeros(len(days))
slp_0_3 = np.zeros(len(days))
clip = 3.5
for d in days:
    slp_1[d] = duration_m(d, 1, clip)
    slp_0_1[d] = duration_m(d, 0.1, clip)
    slp_0_3[d] = duration_m(d, 0.3, clip)
```

```python
plt.figure()
plt.plot(days/365., slp_1, label='Slope=1')
plt.plot(days/365., slp_0_1, label='Slope=0.1')
plt.plot(days/365., slp_0_3, label='Slope=0.3')
plt.legend()
```
